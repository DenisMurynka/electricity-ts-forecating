```python
#combine all files to one master file
import os
import pandas as pd

path = '/home/denis-root/Documents/diploma 2022/py proj/raw data (copy)'

df = []
for file in os.listdir(path):
    if file.endswith('.xlsx'):
        #print('Loading file {0}...'.format(file))
        df.append(pd.read_excel(os.path.join(path,file),sheet_name='Sheet1'))

df_master = pd.concat(df,axis = 0)
df_master.to_excel('masterFile.xlsx', index = False) 
```
